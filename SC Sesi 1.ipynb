```python

user = input("masukkan username anda: ")
pas = input("masukkan password anda: ")
while pas != "Admin123":
  print("password salah")
  pas = input("masukkan password: ")
print("Login Berhasil, Selamat datang ", user)

print("\n")

Name = input("Masukan nama siswa : " )
UAS = float(input("Masukan Nilai UAS : "))
while UAS > 100 or UAS < 0:
  print("Invalid Input")
  UAS = float(input("Masukan Nilai UAS : "))
UTS = float(input("Masukan Nilai UTS : "))
while UTS > 100 or UTS < 0:
  print("Invalid Input")
  UTS = float(input("Masukan Nilai UTS : "))
ASG = float(input("Masukan Nilai Assigment : "))
while ASG > 100 or ASG <0:
  print("Invalid Input")
  ASG = float(input("Masukan Nilai Assigment : "))

Final = (UAS * 0.5) + (UTS * 0.3) + (ASG * 0.2)

if Final <= 11:
  Grade = "F"
elif Final <= 59:
  Grade = "E"
elif Final <= 64:
  Grade = "D"
elif Final <= 69:
  Grade = "C"
elif Final <= 74:
  Grade = "B-"
elif Final <= 79:
  Grade = "B"
elif Final <= 84:
  Grade = "B+"
elif Final <= 89:
  Grade = "A-"
elif Final <= 100:
  Grade = "A"

print("=================================================")
print("Hasil perolehan nilai", Name)
print("Final Score:", Final)
print("Grade ", Grade)
```
