```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f = lambda t, s: np.exp(-t)
h = 0.1
t_min = 0
t_max = 1

t = np.arange(t_min, t_max + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label ='approx (euler h = 0.1)')
plt.plot(t, -np.exp(-t), 'g', label='exact solution')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-v0_8-poster')

def system_ode_rhs(x, Y):
    y, y_prime = Y
    dy_dx = y_prime
    dy_prime_dx = x**2 - 4 * y
    return np.array([dy_dx, dy_prime_dx])

h = 0.1

x_min = 0
x_max = 5

t = np.arange(x_min, x_max + h, h)

s0 = np.array([1.0, 0.0])

s = np.zeros((len(t), 2))
s[0] = s0

for i in range(0, len(t) - 1):
    delta_Y = h * system_ode_rhs(t[i], s[i])
    s[i + 1] = s[i] + delta_Y

y_approx = s[:, 0]
y_prime_approx = s[:, 1]

def exact_solution(x):
    return (9/8) * np.cos(2 * x) + (1/4) * x**2 - 1/8

y_exact_values = exact_solution(t)

plt.figure(figsize=(12, 8))

plt.plot(t, y_approx, 'o--', label=f'Euler (h = 0.1)')

plt.plot(t, y_exact_values, 'g-', label='Solusi Eksak')

plt.title(r'Solusi $y\'\' + 4y = x^2$ dengan $y(0)=1, y\'(0)=0$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn-v0_8-poster')

f = lambda x, y: np.sin(y)

h = 0.1
x_min = 0
x_max = 1

x = np.arange(x_min, x_max + h, h)
y0 = 1
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

plt.figure(figsize=(12,8))
plt.plot(x, y, 'bo--', label="approx (Euler's method, h=0.1)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Euler's Method for y' = sin(y), y(0) = 1")
plt.grid(True)
plt.show()
```
